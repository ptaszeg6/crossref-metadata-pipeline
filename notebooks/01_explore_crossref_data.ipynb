```python
bronze_files = list(Path("../data/bronze").glob("bronze_crossref_works_*.parquet"))
bronze_path = bronze_files[-1]

df = pl.read_parquet(bronze_path)
df.head()
```

```python
profile = []

for col in df.columns:
    profile.append({
        "column": col,
        "dtype": str(df[col].dtype),
        "nulls": df[col].null_count(),
        "null_percent": round(df[col].null_count() / df.height * 100, 2),
        "unique_values": df[col].n_unique(),
    })

pl.DataFrame(profile)
```

```python
author_keys = {}

for item in items:
    for author in item.get("author", []):
        if isinstance(author, dict):
            for key in author.keys():
                author_keys[key] = author_keys.get(key, 0) + 1

sorted(author_keys.items(), key=lambda x: x[1], reverse=True)
```

```python
total_items = len(items)

with_authors = sum(1 for item in items if item.get("author"))
without_authors = total_items - with_authors

with_authors, without_authors
```

```python
author_counts = [
    {
        "doi": item.get("DOI"),
        "title": item.get("title", [None])[0] if item.get("title") else None,
        "author_count": len(item.get("author", [])),
    }
    for item in items
]

author_counts_df = pl.DataFrame(author_counts)
author_counts_df.sort("author_count", descending=True).head(10)
```

```python
author_rows = []

for item in items:
    doi = item.get("DOI")

    for author in item.get("author", []):
        if isinstance(author, dict):
            author_rows.append({
                "doi": doi,
                "given": author.get("given"),
                "family": author.get("family"),
                "orcid": author.get("ORCID"),
                "affiliation": author.get("affiliation"),
            })

authors_df = pl.DataFrame(author_rows)

authors_df.head()
```

```python
from collections import Counter
import json
from pathlib import Path

#raw_path = Path("../data/raw/crossref_raw_20260614T213935Z_fed79583-82ce-45d7-82fa-6d834d0895d6.json")

with raw_path.open("r", encoding="utf-8") as file:
    payload = json.load(file)

items = payload["message"]["items"]
```

```python
field = "title"
#field = "container-title"

type_counts = Counter(type(item.get(field)).__name__ for item in items)

type_counts
```

```python
fields = [
    "title",
    "container-title",
    "short-container-title",
    "original-title",
]

for field in fields:
    type_counts = Counter(type(item.get(field)).__name__ for item in items)
    print(field, type_counts)
```

```python
field = "title"

length_counts = Counter(
    len(item[field])
    for item in items
    if isinstance(item.get(field), list)
)

length_counts
```

```python
field = "original-title"

multi_value_examples = [
    item.get(field)
    for item in items
    if isinstance(item.get(field), list) and len(item[field]) > 1
]

multi_value_examples[:5]
```

```python
field = "title"

missing_count = sum(1 for item in items if field not in item)
empty_list_count = sum(1 for item in items if item.get(field) == [])

missing_count, empty_list_count
```

```python
doi_values = [
    item.get("DOI")
    for item in items
    if item.get("DOI") is not None
]

doi_checks = []

for doi in doi_values:
    doi_checks.append({
        "original": doi,
        "has_outer_spaces": doi != doi.strip(),
        "has_uppercase": doi != doi.lower(),
        "normalized": doi.strip().lower(),
    })

doi_checks[:10]
```

```python
problematic_dois = [
    row for row in doi_checks
    if row["has_outer_spaces"] or row["has_uppercase"]
]

problematic_dois
```

```python
total_dois = len(doi_values)

uppercase_count = sum(1 for doi in doi_values if doi != doi.lower())
outer_space_count = sum(1 for doi in doi_values if doi != doi.strip())

total_dois, uppercase_count, outer_space_count
```
