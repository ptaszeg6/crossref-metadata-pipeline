```python
from pathlib import Path

import polars as pl
import duckdb
```

```python
silver_dir = Path("../data/silver")
bronze_dir = Path("../data/bronze")

publications_files = sorted(silver_dir.glob("silver_publications_*.parquet"))
authors_files = sorted(silver_dir.glob("silver_publication_authors_*.parquet"))
identifiers_files = sorted(silver_dir.glob("silver_publication_identifiers_*.parquet"))
quality_files = sorted(silver_dir.glob("silver_quality_issues_*.parquet"))
bronze_files = sorted(bronze_dir.glob("bronze_*.parquet"))

publications_path = publications_files[-1]
authors_path = authors_files[-1]
identifiers_path = identifiers_files[-1]
quality_path = quality_files[-1]
bronze_path = bronze_files[-1]

publications_path, authors_path, identifiers_path, quality_path, bronze_path
```

```python
publications_df = pl.read_parquet(publications_path)
authors_df = pl.read_parquet(authors_path)
identifiers_df = pl.read_parquet(identifiers_path)
quality_df = pl.read_parquet(quality_path)
bronze_df = pl.read_parquet(bronze_path)

publications_df.shape, authors_df.shape, quality_df.shape, bronze_df.shape, identifiers_df.shape
```

```python
date_comparison = publications_df.select(
    pl.len().alias("total_publications"),

    (
        pl.col("issued_date").is_not_null() &
        pl.col("published_date").is_not_null()
    ).sum().alias("both_dates_present"),

    (
        pl.col("issued_date").is_null() &
        pl.col("published_date").is_null()
    ).sum().alias("both_dates_missing"),

    (
        pl.col("issued_date").is_not_null() &
        pl.col("published_date").is_not_null() &
        (pl.col("issued_date") == pl.col("published_date"))
    ).sum().alias("same_issued_and_published"),

    (
        pl.col("issued_date").is_not_null() &
        pl.col("published_date").is_not_null() &
        (pl.col("issued_date") != pl.col("published_date"))
    ).sum().alias("different_issued_and_published"),
)

date_comparison
```

```python
issued_published_differences = (
    publications_df
    .filter(
        pl.col("issued_date").is_not_null() &
        pl.col("published_date").is_not_null() &
        (pl.col("issued_date") != pl.col("published_date"))
    )
    .select(
        "doi",
        "title",
        "published_date",
        "issued_date",
        "published_print_date",
        "published_online_date",
    )
)

issued_published_differences
```

```python
#check DOI uniqueness in silver publications
duplicate_dois = (
    publications_df
    .group_by("doi")
    .len()
    .filter(pl.col("len") > 1)
)

duplicate_dois
```

```python
#check important missing fields
publications_df.select(
    pl.len().alias("total_publications"),
    pl.col("doi").is_null().sum().alias("missing_doi"),
    pl.col("title").is_null().sum().alias("missing_title"),
    pl.col("published_date").is_null().sum().alias("missing_published_date"),
    pl.col("publisher").is_null().sum().alias("missing_publisher"),
    pl.col("container_title").is_null().sum().alias("missing_container_title"),
)
```

```python
#optional field coverage
publications_df.select(
    pl.len().alias("total_publications"),
    pl.col("abstract").is_not_null().sum().alias("has_abstract"),
    pl.col("language").is_not_null().sum().alias("has_language"),
    pl.col("article_number").is_not_null().sum().alias("has_article_number"),
    (pl.col("license_json") != "[]").sum().alias("has_license"),
    (pl.col("isbn_json") != "[]").sum().alias("has_isbn"),
    (pl.col("alternative_ids_json") != "[]").sum().alias("has_alternative_ids"),
)
```

```python
#publication type distribution
(
    publications_df
    .group_by("publication_type")
    .len()
    .sort("len", descending=True)
)
```

```python
#top publishers
(
    publications_df
    .group_by("publisher")
    .len()
    .sort("len", descending=True)
    .head(20)
)
```

```python
#author sanity checks
authors_df.select(
    pl.len().alias("total_author_rows"),
    pl.col("doi").is_null().sum().alias("missing_doi"),
    pl.col("given_name").is_null().sum().alias("missing_given_name"),
    pl.col("family_name").is_null().sum().alias("missing_family_name"),
    pl.col("orcid").is_not_null().sum().alias("authors_with_orcid"),
)
```

```python
#authors per publication
authors_per_publication = (
    authors_df
    .group_by("doi")
    .len()
    .rename({"len": "author_count"})
    .sort("author_count", descending=True)
)

authors_per_publication.head(20)
```

```python
#publications without authors
publications_without_authors = (
    publications_df
    .join(
        authors_per_publication,
        on="doi",
        how="left",
    )
    .filter(pl.col("author_count").is_null())
    .select("doi", "title", "publisher", "publication_type")
)

publications_without_authors
```

```python
# quality issues summary
(
    quality_df
    .group_by("issue_type")
    .len()
    .sort("len", descending=True)
)
```

```python
#inspect future publication dates
quality_df.filter(
    pl.col("issue_type") == "future_published_date"
)
```

```python
#Then check corresponding publications:
future_dois = (
    quality_df
    .filter(pl.col("issue_type") == "future_published_date")
    .select("doi")
)

publications_df.join(future_dois, on="doi", how="inner").select(
    "doi",
    "title",
    "published_date",
    "published_print_date",
    "published_online_date",
    "issued_date",
)
```

```python
# DuckDB check
#This is nice because your silver layer is Parquet and should be queryable.
duckdb.sql(f"""
    SELECT
        publication_type,
        COUNT(*) AS publication_count
    FROM '{publications_path.as_posix()}'
    GROUP BY publication_type
    ORDER BY publication_count DESC
""").df()
```

```python
#final short summary
summary = {
    "publications": publications_df.height,
    "authors": authors_df.height,
    "quality_issues": quality_df.height,
    "duplicate_dois": duplicate_dois.height,
    "publications_without_authors": publications_without_authors.height,
}

summary
```
